```python
import csv
with open('parent.csv', mode = 'r', encoding = 'utf-8') as f:
    student_data = csv.reader(f)

    try:
        header = next(student_data)
    except StopIteration:
        print("The CSV is empty")
        header = []
    if header:
        header = [h for h in header] + ['student_id'] 
        modified_data = [header]

        # 逐行處理資料
        student_id_prefix = "s_"  # 學生 ID 的前綴
        student_id_sem = ["2021", "2022", "2023"]  # 學年度
        student_id_counter = 1  # 用於生成學生 ID 的計數器

        year_index = 0  # 用來選擇學年度
        rows_per_year = 10000  # 每個學年度有 10,000 條數據

        for row_index, row in enumerate(student_data, start=1):
            # 每過 10,000 條數據，就切換學年度
            if row_index > rows_per_year * (year_index + 1):
                year_index += 1
            
            # 根據計數器生成 student_id
            year = student_id_sem[year_index]  # 根據 year_index 選擇學年度
            student_id = f"{student_id_prefix}{year}{str(student_id_counter).zfill(5)}"

            # 在每一行的最前面插入 student_id
            row.insert(2, student_id)
            # 把修改過的行資料加入到新列表中
            modified_data.append(row)

            # 更新計數器
            student_id_counter += 1

# 將修改後的資料寫回到新的 CSV 檔案
with open('parent_data2.csv', mode='w', newline='', encoding='utf-8') as outfile:
    writer = csv.writer(outfile)
    writer.writerows(modified_data)

    
    
```

```python


# 讀取原始的 CSV 檔案
with open('modified_student_data2.csv', mode='r', newline="", encoding='utf-8') as infile:
    reader = csv.reader(infile)
    
    # 讀取標頭
    try:
        header = next(reader)
    except StopIteration:
        print("The CSV file is empty.")
        header = []

    if header:
        # 在標頭最前面加上 'student_id' 欄位，並刪除 "city", "district", "route" 欄位，新增 "school", "grade", "tel", "status"
        modified_header = ['student_id'] + [h for h in header]

        # 儲存所有處理過的資料
        modified_data = [modified_header]

        # 逐行處理資料
        student_id_prefix = "s_"  # 學生 ID 的前綴
        student_id_sem = ["2021", "2022", "2023"]  # 學年度
        student_id_counter = 1  # 用於生成學生 ID 的計數器

        year_index = 0  # 用來選擇學年度
        rows_per_year = 10000  # 每個學年度有 10,000 條數據

        for row_index, row in enumerate(reader, start=1):
            # 每過 10,000 條數據，就切換學年度
            if row_index > rows_per_year * (year_index + 1):
                year_index += 1
            
            # 根據計數器生成 student_id
            year = student_id_sem[year_index]  # 根據 year_index 選擇學年度
            student_id = f"{student_id_prefix}{year}{str(student_id_counter).zfill(5)}"

            # 在每一行的最前面插入 student_id
            row.insert(0, student_id)
            # 把修改過的行資料加入到新列表中
            modified_data.append(row)

            # 更新計數器
            student_id_counter += 1

# 將修改後的資料寫回到新的 CSV 檔案
with open('student_data5.csv', mode='w', newline='', encoding='utf-8') as outfile:
    writer = csv.writer(outfile)
    writer.writerows(modified_data)
```
